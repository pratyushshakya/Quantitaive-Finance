```python
#Now Calculate the stdev of daily return 

aapl['Daily Return'].std()
```

```python
#Now we will combine everything into a function that will calculate the daily sharpe ratio
    #Compute mean daily return, calculate stdev and use ratio between them and risk free rate to return sharpe ratio
    
    
def compute_sharpe_ratio(df,risk_free_rate = 0):
    
    #First get mean return
    mean_return = df['Daily Return'].mean()
    
    #Now get stdev
    std = df['Daily Return'].std()
    
    #Now calculate the sharpe ratio
    sharpe_ratio = (mean_return - risk_free_rate)/std
    
    return sharpe_ratio #This will return daily sharpe ratio since we are calculating everything from daily return
    
    
```

```python
sr_apple = compute_sharpe_ratio(aapl)
```

```python
sr_apple
```

```python
#Now we can have the function annulalize the sharpe ratio as well

def annual_sharpe_ratio(df, rf = 0):

    mean_return = df['Daily Return'].mean()
    
    stdev = df['Daily Return'].std()
    
    annual_sharpe_ratio = (mean_return - rf)/ stdev
    
    return annual_sharpe_ratio * (252**0.5) #alternatively np.sqrt(252)
```

```python
annual_sr_apple = annual_sharpe_ratio(aapl)
```

```python
annual_sr_apple
```

```python
annual_sr_msft = annual_sharpe_ratio(msft)
```
